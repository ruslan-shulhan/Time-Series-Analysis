**Objective**

```python
# import libraries
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (12, 6))
sns.set(rc={'figure.figsize':(11.7,7.27)})
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

# stat tools
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima

# metrics
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse

from sklearn.preprocessing import MinMaxScaler

#from keras_preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

from fbprophet import Prophet
```

```python
# loading dataset
path = "/content/drive/MyDrive/Colab Notebooks/data/4/DOGE-USD.csv"
# path = "/content/drive/MyDrive/Colab Notebooks/data/4/ETH-USD.csv"
data = pd.read_csv(path)
data
```

```python
# removing unrelated columns
df = data[['Date', 'Close']]
df.head()
```

```python
# checking for datatypes
df.info()
```

```python
# converting 'Date' column into 'datetime' type
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.dtypes
```

```python
# checking for missing values
df.isnull().sum()
```

```python
mask_1 = df['Close'].isnull()
df[mask_1]
```

```python
# filling missing values using: 'ffill' method
df['Close'] = df['Close'].fillna(method='ffill')
df.isnull().sum()
```

```python
# filtring infromation for 2021 year only
mask_2 = (df['Date'] > "2020-12-31") & (df['Date'] < "2022-01-01")
df = df[mask_2]
df
```

```python
# setting 'Date' column as index for graphical visualization
df_idx = df.set_index(['Date'])
df_idx
```

```python
# downsampling from days to months for simplicity
# df_idx_grp = df_idx.groupby(pd.Grouper(freq='D')).mean()
# df_idx_grp = df_idx_grp.sort_index()
# df_idx_grp.shape
df_idx_grp = df_idx
```

```python
# plotting graph
_ = sns.lineplot(data=df_idx_grp)
_ = plt.ylabel('Close price: USD')
_ = plt.title('Line plot - Stock price for DogeCoin 2021')
plt.show()
```

```python
# applying Seasonal decomposition using moving averages
# https://www.statsmodels.org/stable/generated/statsmodels.tsa.seasonal.seasonal_decompose.html
seas_components = seasonal_decompose(df_idx_grp['Close'], model='multiplicative')
```

```python
seas_components.plot()
plt.show()
```

```python
# let's see seasonality of our data closer
_ = plt.plot(seas_components.seasonal)
_ = plt.title("Seasonality of 'Dogecoin' price")
```

**Forecasting using: Prophet model**

```python
# reseting index because Prohet model will require two specific named columns
df_pr = df_idx_grp.reset_index()
df_pr
```

```python
# renaming columns
df_pr = df_pr.rename(columns={'Date': 'ds', 'Close': 'y'})
df_pr.columns
```

```python
# splitting data into training/testing parts
train_data_pr = df_pr[:len(df_pr)-15]
test_data_pr = df_pr[len(df_pr)-15:]
```

```python
# creating instance of model Prophet
model_1 = Prophet()
model_1.fit(train_data_pr)
```

```python
# getting prediction values for 2 weeks
future = model_1.make_future_dataframe(periods=15, freq="D")
model_1_pred = model_1.predict(future)
```

```python
# taking a look at results
model_1_pred.tail()
```

```python
# filtering results of predictions
model_1_pred_final = model_1_pred[['ds', 'yhat']]
mask_3 = model_1_pred_final['ds'] > '2021-12-16'
model_1_pred_final = model_1_pred_final[mask_3]
model_1_pred_final
```

```python
# identifying the metrics between predicted values and original ones
def get_error_metrics(data):
  rmse_error = rmse(test_data_pr['y'], data)
  mse_error = rmse_error ** 2
  mean_value = df_idx_grp['Close'].mean()

  print(f"RMSE Error: {rmse_error}")
  print(f"MSE Error: {mse_error}")
  print(f"Mean Value: {mean_value}")

  return {
      "RMSE Error": rmse_error,
      "MSE Error": mse_error
      }
```

```python
# checking summary metrics
model_1_error_results = get_error_metrics(model_1_pred_final['yhat'])
```

```python
# visualizing results
_ = plt.plot(test_data_pr['ds'], test_data_pr['y'], label='Original')
_ = plt.plot(model_1_pred_final['ds'], model_1_pred_final['yhat'], label='Prediction')
_ = plt.legend(loc='best')
_ = plt.title('Summary of Dogecoin prediction using Prophet model')
_ = plt.ylabel('Stock cost - (USB)')
_ = plt.xlabel('Date')
plt.show()
```

```python
# applying to whole dataset
# creating instance of model Prophet
model_1_final = Prophet()
model_1_final.fit(df_pr)

# getting prediction values for 2 weeks
future_final = model_1_final.make_future_dataframe(periods=15, freq="D")
model_1_pred_final = model_1_final.predict(future_final)

# taking a look at results
print(model_1_pred_final.tail())

# filtering results of predictions
model_1_pred_final = model_1_pred_final[['ds', 'yhat']]
mask_4 = model_1_pred_final['ds'] > '2021-12-31'
model_1_pred_final = model_1_pred_final[mask_4]
model_1_pred_final = model_1_pred_final.rename(columns={'ds':'Date',
                                                       'yhat':'Prediction'})
print(model_1_pred_final)

# visualizing results
_ = plt.plot(model_1_pred_final['Date'], model_1_pred_final['Prediction'], label='Prediction')
_ = plt.legend(loc='best')
_ = plt.title('Summary of Dogecoin prediction using Prophet model for 2022 \
(first two weeks of Jan)')
_ = plt.ylabel('Stock cost - (USB)')
_ = plt.xlabel('Date')
plt.show()
```
